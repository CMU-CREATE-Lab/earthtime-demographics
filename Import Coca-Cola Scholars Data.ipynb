```python
nces_cs_hs_map={}

hs_specs_per_cs = nces_hs_2015[['CITY_STATE','HS_SPEC']].groupby('CITY_STATE')
```

```python
for cs,group in hs_specs_per_cs:
    print "%s %d" % (cs, len(group))
    hs_set = set()
    for i in range(0,len(group)):
        hs = group['HS_SPEC'].iloc[i]
        print "%d: %s"%(i,hs)
        hs_set.add(hs)
    nces_cs_hs_map[cs]=hs_set
```

```python
# Deal with the issue that some cities are included as both all uppercase and not uppercase
for cs in nces_cs_hs_map:
    cs_u = cs.upper()
    if(cs != cs_u and cs_u in nces_cs_hs_map):
        all_hs = nces_cs_hs_map[cs].union(nces_cs_hs_map[cs_u])
        print "Merging %s (%d) and %s (%d), total = %d"%(cs,len(nces_cs_hs_map[cs]),cs_u,len(nces_cs_hs_map[cs_u]),len(all_hs))
        nces_cs_hs_map[cs]= all_hs
        nces_cs_hs_map[cs_u]=all_hs
        
```

```python
for i in range(0,len(ccsd_2018)):
    cs = ccsd_2018['CITY_STATE'].iloc[i]
    hs = ccsd_2018['HS_SPEC'].iloc[i]
    print "%d: %s %s"%(i,cs,hs)
    if(hs in hs_remap):
        print "  OK (hs_remap)"
        continue
    elif(hs in hs_latlon_map):
        print "  OK (%f,%f)"%(hs_latlon_map[hs])
        continue

    if cs not in nces_cs_hs_map:
        if(cs.upper() in nces_cs_hs_map):
            # Check if upcased cs is in there
            cs=cs.upper()
        elif(cs.capitalize() in nces_cs_hs_map):
            cs=cs.capitalize()
        else:
            cs_not_in_nces.add(cs)
            print "  NO matching CITY_STATE"
            continue
    if hs not in nces_cs_hs_map[cs]:
        if(hs.upper() in nces_cs_hs_map[cs]):
            print "  OK (%s)"%(hs.upper())
        elif(re.sub(cs,cs.upper(),hs) in nces_cs_hs_map[cs]):
            # Check if upcased cs is in there
            print "  OK (%s->%s)"%(cs,cs.upper())
        elif(re.sub(cs,cs.capitalize(),hs) in nces_cs_hs_map[cs]):
            print "  OK (%s->%s)"%(cs,cs.capitalize())
        elif(hs.replace("School","Sch") in nces_cs_hs_map[cs]):
            # Check if School/Sch
            print "  %r:%r"%(hs,hs.replace("School","Sch"))
            hs_remap[hs]=hs.replace("School","Sch")
        elif(hs.replace(" School","") in nces_cs_hs_map[cs]):
            print "  %r:%r"%(hs,hs.replace(" School",""))
            hs_remap[hs]=hs.replace(" School","")
        elif(hs.replace("High School","Senior High") in nces_cs_hs_map[cs]):
            print "  %r:%r"%(hs,hs.replace("High School","Senior High"))
            hs_remap[hs]=hs.replace("High School","Senior High")
        elif(re.sub(r' ([A-Z]) ',r' \1. ',hs) in nces_cs_hs_map[cs]):
            repl_hs = re.sub(r' ([A-Z]) ',r' \1. ',hs)
            print "  %r:%r"%(hs,repl_hs)
            hs_remap[hs]=repl_hs
        elif(len(nces_cs_hs_map[cs])==1):
            print "  ONLY %r:%r"%(hs,list(nces_cs_hs_map[cs])[0])
            hs_remap[hs]=list(nces_cs_hs_map[cs])[0]
        elif('Home School' not in hs):
            hs_not_in_nces.add(hs)
            print "  %s not in"%(hs)
            for nces_hs in sorted(nces_cs_hs_map[cs]):
                print "    %s"%(nces_hs)
    else:
        print "  EXACT"
        hs_exact_in_nces.add(hs)
```
